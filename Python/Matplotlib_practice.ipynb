
### 2.  Histograms

it was introduced by *Karl Pearson*. It is a representation of numerical data in a form of bar graphs. Following is a bried explaination: -

* BINS: these are nothing division of entire range of values into some specific series of intervals. These intervals are non overlapping. 
* Then we plot the count of values or frequency of values which fall in these bins or buckets in form of a bar


```python
iris.plot.hist(figsize = (20,6), bins = 15)  # iris.plot(kind = 'hist')
```

```python
iris['sepal_length'].hist()
```

#### you can also specifiy unequally spaced bins by passing a list
in the below example one bin is between 1 and 4 and other one is 
between 4 and 10
* first bin excludes 4 $i.e$ its range $\in\ [1,4)$ 
* second bin or last bin includes both 5 and 10 both $i.e.$ its in range $\in\ [4,10]$

```python
iris['sepal_length'].hist(bins = [1,4,10])
```

```python
# here we have plotted stacked bar graph with bins = 50
# to observe difference between stacked and unstacked lets plot 
# both the cases
# case 1: stacked = True {default case}
 
iris.plot(kind = 'hist', stacked = True, bins = 50)
```

```python
# case 2: stacked = False

iris.plot(kind = 'hist', stacked = False, bins = 50)
```

```python
# we can also make stacked horizontal historgrams
# by using orientaion = 'horizontal' parameter

iris.plot(kind = 'hist', stacked = True, bins = 50, orientation = 'horizontal', figsize = (10,10))
```

```python
# Lets calculate the first discrete difference of element
# Its nothing but substracting current cell or row from previous cell
iris['sepal_width'].diff()[:5]
```

```python
iris['sepal_width'].diff().plot(kind = 'hist', stacked =True, bins = 50)
```

### 7. 3D plot

```python
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X,Y,Z, 50, cmap = 'binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_wireframe(X,Y,Z, cmap = 'flare')
ax.set_title('wireframe')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set_title('surface');
```

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)
ax = plt.axes(projection='3d')
ax.scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5);
```
