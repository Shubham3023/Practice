````python
import pandas as pd

URL = "https://archive.ics.uci.edu/ml/machine-learning-databases/yeast/yeast.data"
yeast_df = pd.read_table(URL,sep="\s+", header=None)
yeast_df.columns =['Sequence Name','mcg', 'gvh', 'alm', 'mit', 'erl', 'pox','vac', 'nuc',  'class_protein_localization' ]
yeast_df.head()
````

````python
cols = yeast_df.columns[1:-1]
for i in cols:
    plt.hist(yeast_df[i], density = True, alpha = 0.6)
````

````python
sns.distplot(yeast_df.mcg)
````

````python
sns.distplot(yeast_df.alm)
````

````python
numerical_yeast_df = yeast_df.iloc[:, 1:-1]
numerical_yeast_df.head()
````

````python
sns.kdeplot(data = numerical_yeast_df ,x = 'mcg', y =  'alm')
````

````python
with sns.axes_style('ticks'):
    sns.jointplot(x = 'mcg', y = 'alm', data = yeast_df, kind = 'kde', fill = True)
````

````python
with sns.axes_style('ticks'):
    sns.jointplot(x = 'mcg', y = 'alm', data = yeast_df, kind = 'hex', gridsize = 20 );
````

````python
yeast_df.info()
````

````python
sns.pairplot(data = yeast_df, hue='class_protein_localization', height= 3)
````

````python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
sns.set(style = 'darkgrid')
````

````python
tips = sns.load_dataset('tips')
tips.head()
````


### 1.1 ```relplot()```
**Full signature of relplot**-
```python

seaborn.relplot(x=None, y=None, hue=None, size=None, style=None, data=None, row=None, col=None, col_wrap=None, row_order=None, col_order=None, palette=None, hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, markers=None, dashes=None, style_order=None, legend='brief', kind='scatter', height=5, aspect=1, facet_kws=None, **kwargs)
```

````python
sns.relplot(x = 'total_bill' , y = 'tip', data= tips)
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style= 'size')
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style= 'size')
````

````python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size', palette= 'ch:r=-0.8, l = 0.95')
````

````python
# size = 'size' can be used to manipulate diameter of circle for better understanding of data
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size', hue = 'smoker') 
````

````python
# we can also specify size range for size keyword in tuple format
# it maps with size column value 
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size', sizes = (15,200), hue = 'smoker') 
````

````python
data = dict(time = np.arange(500),
           value = np.random.randn(500).cumsum())

df = pd.DataFrame(data)
df.head()
````

````python
# plotting the data and this time changing kind to line.
sns.relplot(x = 'time', y = 'value', data = df, kind = 'line', sort = True)
````

````python
dummy_data = np.random.randn(500,2).cumsum(axis = 0)
df1 = pd.DataFrame(dummy_data, columns=['time', 'value'])
df1.head()
````

````python
# whene sorting is false it plots a random ouput
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df1, sort = False)
````

````python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df1, sort = True)
````
