```python
df_sales = pd.read_csv('sales_data_final.csv')
```

```python
df_sales.head()
```

```python
#apply a function to data
```

```python
def profit_flag(a):
    if a >= 0:
        return 'Positive'
    else:
        return 'Negative'
```

```python
df_sales['profit_flag'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
def len_customer(a) :
    return len(a)
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len_customer)

#also below can be used directly
# df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a) :
    if a < 5:
        return "low"
    elif a >= 5 and a < 9:
        return "medium"
    else:
        return "high"
```

```python
df_sales['quantity_flag'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.tail()
```

```python
df_sales['quantity_squared'] = df_sales['quantity'].apply(lambda x : x **2)
```
